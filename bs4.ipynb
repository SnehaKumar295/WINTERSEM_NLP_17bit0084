```python
import bs4
```

```python
import requests
```

```python
k=requests.get("https://www.indiatoday.in/news.html");
```

```python
l=bs4.BeautifulSoup(k.content,'html5lib');
```

```python
print(l.title.string)
```

```python
print(l.prettify())
```

```python
x=l.find_all("a");
```

```python
for links in x:
    print(links.get('href'));
```

```python
p=l.get_text();
```
